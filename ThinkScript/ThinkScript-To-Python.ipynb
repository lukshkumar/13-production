```python
def Sum_Thinkorswim(input_list, total_values_to_iterate, starting_index = 0):
    output_list = []
    for i in range(len(input_list)):
        #Validation for previous data on 0th index (First value)
        if(starting_index >= 0):
            range_start = (i - starting_index) >= 0 ? (i - starting_index) : 0
            range_end = (range_start - total_values_to_iterate) >= -1 ? (range_start - total_values_to_iterate) : -1 
        partial_sum = 0
        for j in range(range_start, range_end, -1):
            partial_sum += input_list[j]
            
        output_list.append(partial_sum)
    return output_list

def Average_Thinkorswim(input_list, total_values_to_iterate, starting_index = 0):
    output_list = []
    for i in range(len(input_list)):
        #Validation for previous data on 0th index (First value)
        if(starting_index >= 0):
            range_start = (i - starting_index) >= 0 ? (i - starting_index) : 0
            range_end = (range_start - total_values_to_iterate) >= -1 ? (range_start - total_values_to_iterate) : -1 
        partial_sum = 0
        total_values_to_average = 0
        for j in range(range_start, range_end, -1):
            partial_sum += input_list[j]
            total_values_to_average += 1
        
        partial_average = partial_sum/total_values_to_average 
        output_list.append(partial_average)
    return output_list
            
```
